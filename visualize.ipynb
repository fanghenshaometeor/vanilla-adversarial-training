```python
""" load data and print shape """
""" """
import numpy as np 

pca_train = np.load('log/pca_train.npy')
pca_test = np.load('log/pca_test.npy')
pca_train_FGSM = np.load('log/pca_train_FGSM.npy')
pca_test_FGSM = np.load('log/pca_test_FGSM.npy')

pca_tr_features, pca_tr_labels = pca_train[:,0:2], pca_train[:,2].astype(np.uint8)
pca_te_features, pca_te_labels = pca_test[:,0:2], pca_test[:,2].astype(np.uint8)

pca_tr_f_FGSM, pca_tr_l_FGSM = pca_train_FGSM[:,0:2], pca_train_FGSM[:,2].astype(np.uint8)
pca_te_f_FGSM, pca_te_l_FGSM = pca_test_FGSM[:,0:2], pca_test_FGSM[:,2].astype(np.uint8)


# print(pca_tr_features.shape)
# print(pca_tr_labels.shape)
# print(pca_te_features.shape)
# print(pca_te_labels.shape)

# print(pca_tr_labels.max())
# print(pca_tr_labels.min())
# print(pca_te_labels.max())
# print(pca_te_labels.min())
```

```python
""" pca plot """
""" """
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

fig = plt.figure(1)
plt.xlim(-5,9)
plt.ylim(-7,7)
for i in range(10):
    pca_tr_features_sub = pca_tr_features[np.where(pca_tr_labels==i)]
    print('clean training sample number of label %d = %d'%(i, pca_tr_features_sub.shape[0]))
    plt.scatter(pca_tr_features_sub[:,0], pca_tr_features_sub[:,1])
plt.title('PCA on clean training features')
plt.savefig('./log/pca_train.png')
print('----------------')

fig = plt.figure(2)
plt.xlim(-5,9)
plt.ylim(-7,7)
for i in range(10):
    pca_te_features_sub = pca_te_features[np.where(pca_te_labels==i)]
    print('clean test sample number of label %d = %d'%(i, pca_te_features_sub.shape[0]))
    plt.scatter(pca_te_features_sub[:,0], pca_te_features_sub[:,1])
plt.title('PCA on clean test features')
plt.savefig('./log/pca_test.png')
print('----------------')

fig = plt.figure(3)
plt.xlim(-5,9)
plt.ylim(-7,7)
for i in range(10):
    pca_tr_features_sub = pca_tr_f_FGSM[np.where(pca_tr_l_FGSM==i)]
    print('training adv. example number of label %d = %d'%(i, pca_tr_features_sub.shape[0]))
    plt.scatter(pca_tr_features_sub[:,0], pca_tr_features_sub[:,1])
plt.title('PCA on FGSM training features')
plt.savefig('./log/pca_train_FGSM.png')
print('----------------')

fig = plt.figure(4)
plt.xlim(-5,9)
plt.ylim(-7,7)
for i in range(10):
    pca_te_features_sub = pca_te_f_FGSM[np.where(pca_te_l_FGSM==i)]
    print('test adv. example number of label %d = %d'%(i, pca_te_features_sub.shape[0]))
    plt.scatter(pca_te_features_sub[:,0], pca_te_features_sub[:,1])
plt.title('PCA on FGSM test features')
plt.savefig('./log/pca_test_FGSM.png')
print('----------------')
```
